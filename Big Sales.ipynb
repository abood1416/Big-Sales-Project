# **Define y(dependent variable) and X(independent feature)**

```python
y= df['Item_Outlet_Sales']
```

```python
y.shape
```

```python
y
```

```python
X= df.drop(['Item_Identifier','Item_Outlet_Sales'],axis=1)
```

```python
X.shape
```

```python
X
```

# **Get X Variables Standardized**

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_std = df[['Item_Weight','Item_Visibility', 'Item_MRP','Outlet_Establishment_Year']]
```

```python
X_std = sc.fit_transform(X_std)
```

```python
X_std
```

```python
X[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']] = pd.DataFrame(X_std,columns=[['Item Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']])
```

```python
X
```

# **Get Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1, random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# **Get Model Train**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(random_state=2529)
```

```python
rfr.fit(X_train, y_train)
```

# **Get Model Prediction**

```python
y_pred = rfr.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
y_pred
```

# **Get Model Evaluation**

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

# **Get Visualization of Actual VS predicted Results**

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
